```python
# student=[{"name":"Aman","marks":90,"lab_assign":True,"city":"Agra","sec":"CB"},{"name":"Anubhav","marks":90,"lab_assign":True,"city":"Mathura","sec":"CB"},{"name":"Bhanu","marks":92,"lab_assign":True,"city":"Mathura","sec":"CB"},{"name":"Abhishek","marks":95,"lab_assign":True,"city":"Agra","sec":"CB"}]

students = [
    {"name": "Aman", "marks": 92, "attendance": 95, "lab_assignment": True},
    {"name": "Akshat", "marks": 75, "attendance": 85, "lab_assignment": True},
    {"name": "Anubhav", "marks": 58, "attendance": 60, "lab_assignment": True},
    {"name": "Abhishek", "marks": 88, "attendance": 90, "lab_assignment": False},
    {"name": "Bhanu", "marks": 35, "attendance": 80, "lab_assignment": True}
]

for student in students:
    if student["lab_assignment"] == False:
        result = "Lab Assignment Not Submitted"
    elif student["attendance"] < 75:
        result = "Low Attendance"
    elif student["marks"] >= 70:
        result = "Distinction"
    elif student["marks"] >= 40:
        result = "Pass"
    else:
        result = "Fail"

    print(student["name"], "-", result)
```
